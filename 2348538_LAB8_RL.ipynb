```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_states = 5
start_state = 0
terminal_state = n_states - 1
rewards = np.array([-1, -1, -1, -1, 0])
```

```python
alpha = 0.1
gamma = 0.9
n_episodes = 100
```

```python
V = np.zeros(n_states)

for episode in range(n_episodes):
    state = start_state
    while state != terminal_state:
        next_state = state + 1 if state < terminal_state else state 
        V[state] = alpha * (rewards[next_state] + gamma * V[next_state] - V[state])
        state = next_state
        
```

```python
print("The optimal value function is:")
for i, value in enumerate(V):
    print(f'State {i + 1}: {value: .2f}')
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, n_states + 1), V, marker="o", linestyle="-", color="b")
plt.title("State-Value Estimates (TD(0))")
plt.xlabel("State")
plt.ylabel("Value Estimate")
plt.grid()
plt.show()
```
